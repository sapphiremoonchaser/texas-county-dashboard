# Census API Metadata Explorer

## Browse concepts

```python
# Imports
import pandas as pd

from src.texas_county_dashboards.scripts.census_metadata import (
    CensusVariables
)

# Census object
census = CensusVariables()
```

```python
# Get a uniqe list of all concepts
concepts = census.concepts()

# Turn the series into a list
concepts_list = concepts.tolist()

print(len(concepts_list))
```

```python
print(concepts_list[:5])
```

### Concepts containing 'population' keyword

'Total Population'

```python
keyword = "population"

matching_concepts = [
    concept for concept in concepts
    if keyword.lower() in concept.lower()
]

concepts_df = pd.DataFrame(matching_concepts)

# with pd.option_context("display.max_rows", None):
#     display(concepts_df)
```
